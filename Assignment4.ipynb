```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#load the dataset
df=pd.read_csv('Boston.csv')
df
```

```python
#Removing the unnammed: 15 and unnamed: 16 col
df=df.drop(columns=['Unnamed: 15','Unnamed: 16'])
```

```python
#detecting outliers and removing outliers
```

```python
sns.boxplot(x='CRIM',data=df)
```

```python
#removing outliers z-score
def trim(col:pd.core.series.Series):
    lowerlimit=col.mean()-3*col.std()
    upperlimit=col.mean()+3*col.std()
    newCol= col[(col<upperlimit) & (col>lowerlimit)]
    return newCol
```

```python
#removing outliers for CRIM
sns.boxplot(x=trim(df['CRIM']),data=df)
```

```python
def detect_outliers(df,feature):
    q1=df[feature].quantile(0.25)
    q3=df[feature].quantile(0.75)
    iqr=q3-q1
    lower_lim=q1-1.5*iqr
    upper_lim=q3+1.5*iqr
    return df[(df[feature]<lower_lim)|(df[feature]>upper_lim)]
```

```python
detect_outliers(df,'MEDV')
```

```python
#Regression model 
```

```python
#CHOOSE DEPENDENT AND INDEPENDENT VARIABLES 
```

```python
x=df[['CRIM','LSTAT']]
x.shape
```

```python
y=df['MEDV']
y.shape
```

```python
#Splitting dataset into training and testing set 
from sklearn.model_selection import train_test_split
```

```python
#dataset split into training and testing 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
#create model
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#predicting
y_pred=model.predict(x_test)
```

```python
plt.scatter(y_test,y_pred)
```

```python
#Model Evaluation using mean_absolute_error mean_square_error r2Score
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
meanAbsErr=mean_absolute_error(y_test,y_pred)
meanSqrErr=mean_squared_error(y_test,y_pred)
r2Score=r2_score(y_test,y_pred)
```

```python
print(f"Mean Absolute Error {meanAbsErr}")
print(f"Mean Squared Error {meanSqrErr}")
print(f"r2 Score {r2Score}")
```

```python
#plot graph with bestfit line 
m,b = np.polyfit(y_test,y_pred,1)
```

```python
plt.scatter(y_test,y_pred)
plt.plot(y_test,m*y_test+b,color="orange")
print(f"Slope is m={m}")
print(f" Y-intercept is {b}")
print(f"Equation of line is: y={m}x+ {b}")
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Linear Regression')
plt.grid()
plt.show()
```
