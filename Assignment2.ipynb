```python
#Dealing with outliers
```

```python
def detect_outliers(df,feature):
    q1=df[feature].quantile(0.25)
    q3=df[feature].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    return df[(df[feature]>lower_bound)|(df[feature]<upper_bound)]
    
```

```python
detect_outliers(df,"math score")
```

```python
detect_outliers(df,"reading score")
```

```python
detect_outliers(df,"writing score")
```

```python
#using z score-> standard score -> gives you how far are your values from the mean
```

```python
#handling using z_score
```

```python
def trim(col:pd.core.series.Series):
    lower_limit=col.mean()-3*col.std()
    upper_limit=col.mean()+3*col.std()
    new_col= col[(col<upper_limit)&(col>lower_limit)]
    return new_col
```

```python
plt.subplot(2,2,1)
plt.title("Before Trimming")
plt.boxplot(x="math score",data=df,showmeans=True)
plt.subplot(2,2,2)
plt.title("After trimming")
plt.boxplot(x=trim(df["math score"]),data=df,showmeans=True)
```

```python
plt.subplot(2,2,1)
plt.title("Before Trimming")
plt.boxplot(x="reading score",data=df,showmeans=True)
plt.subplot(2,2,2)
plt.title("After trimming")
plt.boxplot(x=trim(df["reading score"]),data=df,showmeans=True)
```
