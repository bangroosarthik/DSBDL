```python
#POS Tag and Stop words Removal
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
#store all english stopwords inside a set 
stopWords=set(stopwords.words('english'))
```

```python
stopWords
```

```python
for sen in sentence:
    WordSet=word_tokenize(sen)
    filteredWords=[word.lower() for word in WordSet if word.lower() not in stopWords]
    print(f"Sentence without stop words{filteredWords}")
    print(f"Sentence with stop words{WordSet}")
    print("POS Tag: ",nltk.pos_tag(filteredWords))
```

```python
#Stemming and Lemmatization 
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
for word in WordSet:
    print(f"{word} - {stemmer.stem(word)}")
```

```python
#Lemmatization 
from nltk.stem import WordNetLemmatizer 
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for word in WordSet:
    print(f"{word}:{lemmatizer.lemmatize(word)}")
```

```python
#update sentence 
sentence=sentence[:3]
new_sent=["".join(sentence)]
new_sent
```

```python
#Tfidf using TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

def cal_Tfidf(documents):
    #created instance 
    tf_vect=TfidfVectorizer()
    #Tfidf returns a matrix and a feature names using get_feature_names_out()
    tf_matrix=tf_vect.fit_transform(documents)
    features=tf_vect.get_feature_names_out()
    return tf_matrix,features


documents=new_sent
tf_matrix,features=cal_Tfidf(documents)
features,tf_matrix.toarray()
```
